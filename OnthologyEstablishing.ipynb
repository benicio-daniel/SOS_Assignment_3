```python
from rdflib import Graph, Namespace, RDF, RDFS, Literal, URIRef
g = Graph()
```

```python
PROV = Namespace("http://www.w3.org/ns/prov#")
g.bind("prov", PROV)

SCHEMA = Namespace("https://schema.org/")
g.bind("schema", SCHEMA)

CROISSANT = Namespace("https://mlcommons.org/croissant/")
g.bind("croissant", CROISSANT)

MLSO = Namespace("http://w3id.org/mlso/")
g.bind("mlso", MLSO)

me = URIRef("http://example.org/agent/benicio")
g.add((me, RDF.type, PROV.Agent))
g.add((me, RDFS.label, Literal("Student – Group 13")))
```
